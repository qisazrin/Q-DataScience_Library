```python
# Seaborn - Data Visualization Library
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings('ignore')
 

# built-in datasets of seaborn
sb.get_dataset_names()
```

```python
dataset=sb.load_dataset('iris')
dataset
```

```python
#lineplot used to generate a line graph
sb.lineplot(x='sepal_length',y='sepal_width',data=dataset)
plt.title('Iris Graph')
plt.show()
```

```python
# adding style to seaborn     #linestyle '--',':','-'
sb.set_style('darkgrid',{'grid.color':'k','grid.linestyle':':'})    
sb.lineplot(x='sepal_length',y='sepal_width',data=dataset)
plt.title('Iris Graph')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
#change the firgure size
plt.figure(figsize=(10,7))
sb.set_style('darkgrid',{'grid.color':'k','grid.linestyle':':'})    
sb.lineplot(x='sepal_length',y='sepal_width',data=dataset)
plt.title('Iris Graph')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
pokemon = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/pokemon.csv')
pokemon
```

```python
#scatterplot()
sb.scatterplot(x='Defense',y='Attack',data=pokemon)
plt.title('Pokemon Power')
plt.show()
```

```python
pokemon.Generation.value_counts().sort_index()
```

```python
#hue()   # can use to create multicolor in scatter
colors=['#C77773','#8BC773','#73C7B9','#7389C7','#191975','#850707']
sb.scatterplot(x='Defense',y='Attack',hue='Generation',palette=colors,data=pokemon)
plt.title('Pokemon Power')
plt.show()
```

```python
#Lmplot()
# used to generate a scatter plot using a regression line

colors=['#C77773','#8BC773','#73C7B9','#7389C7','#191975','#850707']
sb.lmplot(x='Defense',y='Attack',hue='Generation',palette=colors,data=pokemon)
plt.title('Pokemon Power')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sb.countplot(x='Type 1',data=pokemon)
plt.title('Pokemon Type')

plt.show()
```

```python
pkmn_type_colors = ['#3ae40c',  # Grass
                    '#e52106',  # Fire
                    '#0AE3D6',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#ECF709',  # Dragon
                    '#7A786C',  # Dark
                    '#D3D1C0',  # Steel
                    '#91E8EE',  # Flying
                   ]
plt.figure(figsize=(12,5))
sb.countplot(x='Type 1',palette=pkmn_type_colors ,data=pokemon)
plt.xticks(rotation=45)     # can use rotation for sengetkan perkataan di paksi X
plt.title('Pokemon Type')

plt.show()
```
