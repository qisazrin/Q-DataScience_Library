```python
# Pandas fata science library
import pandas as pd
import numpy as np
print(pd.__version__)
```

```python
# numpy array
arr = np.array([10,20,30,40])
print(arr)
print(type(arr))
```

```python
# Pandas series
# A Series is 1d labeled array
# The S in Series must be in capital letter
 
s = pd.Series([10,20,30,40])
print(type(s))
print(s)
print(s.shape)
 
# In Numpy only numbers
#In Series Numbers + Label
```

```python
# Indexing Slicing and Using Loops
s = pd.Series([10,20,30,40,50,60,70] , index=['val1','val2','val3','val4','val5','val6','val7'])
print(s)
```

```python
#using loops 
for
```

```python
for k,v in s.items():
    print(k,v)
```

```python
for i in enumerate(s):
    print(i)
```

```python
#read_csv()
#used to read a csv file
df=pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Position_Salaries.csv')
df
```

```python
#finding the sum of salary using pandas
salaries=df['Salary']
print(type(salaries))

total_sum=salaries.sum()
print(total_sum)
```

```python
#finding the sum of salary using Numpy
salaries=df['Salary'].to_numpy() #convert to numpy
print(type(salaries))

total_sum=np.sum(salaries)
print(total_sum)
```
