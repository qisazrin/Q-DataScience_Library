```python
# Matplotlib Part - 2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
https://raw.githubusercontent.com/yash240990/Python/master/Grade_Set_1.csv
```

```python
#generate a graph using a dataset
data=pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Grade_Set_1.csv')
data
```

```python
# the data of hours studied and test grade only (x,y)
plt.scatter(data.Hours_Studied,data.Test_Grade,color='green')
plt.plot(data.Hours_Studied,data.Test_Grade,color='blue',label='Time Study')      # if hv this dont forget plt.legend()
plt.xlabel('Hours of study',fontsize=12)
plt.ylabel('Grade Test', fontsize=12)
plt.title('Hours studied VS Test grade')
plt.legend()
plt.show()
```

```python
#xticks & Yticks
from numpy import random
x=random.randint(100,size=10)
y=random.randint(50,size=10)
plt.plot(x,y,color='blue')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Xticks and Yticks')
plt.show()
```

```python
#histrogram
# histrogram shows the distribution/group of data by dividing equally
# for example : if you have the ages of 100 people , a histogram help you see how many people are in their in 20's
# 30s , 40s
# ages = [18,21,22,25,26,30,34,35,36,40,42,45,46,50]
# bins = 5
# divide the range of ages (18-50) into 5 equal sized groups
# 18-24 25-31 32-38 39-45 46-52

ages = [18,21,22,25,26,30,34,35,36,40,42,45,46,50]
plt.hist(ages,bins=5,color='red',edgecolor='black')    # used plt.hist for histrogram # used edgecolor to display line in between
plt.title('age Distribution')
plt.xlabel('Age of buyer')
plt.ylabel('Frecuancy of purchased')
plt.xticks(np.arange(18,51,6))
plt.show()

```

```python
# heatmap
import seaborn as sb
data={ 
    'hours_studied':[2,3,4,5,6,7,8],
    'Marks_scored':[30,40,50,60,70,80,90],
    'Tv_watched':[10,9,8,7,6,5,4] }
df=pd.DataFrame(data)
corr=df.corr()
sb.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
 
colors = ['green', 'red', 'blue', 'purple', 'orange', 'brown']
 
xa=[3,1,1,2,3,4,5,5,5,3]
ya=[1,5,6,7,6,7,6,5,5,1]
 
#6 subplots in 2x3 layout
for i in range(6):
    plt.subplot(2, 3, i + 1)  # 2 rows, 3 columns, position i+1
    plt.plot(xa, ya, color=colors[i])
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.grid(True)
    plt.title(f'Plot {i + 1}')
 
plt.tight_layout()
plt.show()
```
