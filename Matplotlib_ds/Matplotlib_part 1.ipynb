```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt       #important b4 use matplotlib
```

```python
# Line graph
# draws a line from point to point
# plot(x,y)
x=[1,2,3]
y=[4,5,2]
plt.plot(x,y)
#--> show() used to display a graph
plt.show()
```

```python
x=[3,1,1,2,3,4,5,5,5,3]
y=[1,5,6,7,6,7,6,5,5,1]

plt.plot(x,y)
plt.show()
```

```python
x=[5,6,2]
y=[1,9,10]
plt.plot(x,y)
plt.title('my first graph')
plt.xlabel('x Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
# Adding styles to a Graph
# https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
from matplotlib import style
style.available
style.use('ggplot')
x1=[3,1,1,2,3,4,5,5,5,3]
y1=[1,5,6,7,6,7,6,5,5,1]
x2=[4,2,2,3,4,5,6,6,6,4]
y2=[2,6,7,8,7,8,7,6,6,2]
plt.plot(x1,y1,color='pink',linewidth=2,linestyle='dotted',label='!Love you')
plt.plot(x2,y2,color='red',label='Love you')
plt.title('First Product')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html

plt.legend(loc=2)         # use legend to display label '!love you/love you'
plt.grid(True,color='beige')
plt.show()
 
```

```python
# Bar Chart
#bar()

plt.bar([1,3,5,7,9],[5,2,7,8,2],label='Data 1')
plt.bar([2,4,6,8,10],[8,6,2,5,6],label='Data 2')
#plt.bar - Insurance Resources and Information.
#plt.bar is your first and best source for all of the information youre looking for. From general topics to more of what you would expect to find here, plt.bar has it all. We hope you find what you...
plt.title('Bar Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.show()
```

```python
# horizontal bar graph
#barh()

products=['Samsung','Iphone','Huawei','Lenovo','Realme']
quantities=[200,123,452,675,565]
plt.barh(products,quantities)
plt.xlabel('Quantity')
plt.ylabel('Product')
plt.title('Stock Inventory')
plt.show()
```

```python
#scatter plot
#scatter()
x=[3,1,1,2,3,4,5,5,5,3]
y=[1,5,6,7,6,7,6,5,5,1]
colour=['blue','purple','pink','yellow','red','beige','silver','magenta','orange','black',]
plt.scatter(x,y,label='Data Point',marker='v',color=colour)    #marker is to change shape of scatter, can check matplotlib website for keyword
plt.ylabel('Y Axis')
plt.title('Scatter plot')
plt.legend()
plt.show()
```

```python
#pie plot
activities=['reading','cycling','singing','sleeping']
col=['blue','yellow','pink','magenta']
hrs=[2,1,4,12]
plt.pie(hrs,labels=activities,colors=col, startangle=30,explode=(0,0.2,0,0),autopct='%1.1f%%')
#startangle ( used for angle of pie), explode()(used to give effect terkeluar dri pie gitu~)
#autopct='%1.1f%%' used for show percentage in pie
# for name make sure labels with 's'  # and pie plot colors have 's' diff with scatter
plt.title('Activity')
plt.show()
```
