```python
# how much execution time will it take
import time
numbers = list(range(1, 10000))  # 1 2 3 4 5..................9999
n=0
start_time = time.time() # record start time
for i in numbers:
 n += i
end_time = time.time() # record end time
print("Total:", n)
print("Execution Time:", end_time - start_time, "seconds") 
print(f"Execution Time: {end_time - start_time:.4f} seconds") # .4f → means “show 4 digits after the decimal point.
```

```python
# numpy copy() vs view()
# copy()
# a copy will create a new independent array in a different memory location
 
org_arr = np.array([1,2,3,4,5])
copied_arr = org_arr.copy()
print(copied_arr)
 
# if you make changes in copied_arr , it won't affect the original array
 
copied_arr[0] = 100
print(copied_arr)
print(org_arr)
 
# if you make changes in org_arr , it won't affect the copied array
 
org_arr[2] = 300
print(org_arr)
print(copied_arr)
```

```python
num=np.array([1,2,3,4,5]) #1 create copy of the original
num1=num.copy()
print(num1)

num1 [0]= 100 # if make changes in copy(num1), it don't effect the original(num)
print(num1)
print(num)

num[2] = 300 # if you change in original(num), it won't effect the copy array(num1)
print(num)
print(num1)
```

```python
# view()
# a view is the representation of the original array
# it shares the same memory location
 
org_arr = np.array([1,2,3,4,5])
copied_arr = org_arr.view()
print(copied_arr)
 
# if you make changes in copied_arr , it will affect the original array
copied_arr[0] = 100
print(copied_arr)
print(org_arr)
 
# if you make changes in org_arr , it will affect the copied array
 
org_arr[2] = 300
print(org_arr)
print(copied_arr)
```

```python
arr = np.array([10, 32, 30, 50, 20, 82, 91, 45])
 
# Find indices where value is 30
indices = np.where(arr == 30)
print(f"Indices where value is 30: {indices}")
 
# Find indices where values are even
even_indices = np.where(arr % 2 == 0)
print(f"Indices of even values: {even_indices}")
 
# Conditional selection
new_arr = np.where(arr > 40, arr * 2, arr)
print(f"Array after conditional selection: {new_arr}")
```
